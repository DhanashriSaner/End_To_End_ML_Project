```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
warnings.filterwarnings("ignore")
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("exploratory_data.csv")
df.head()
```

```python
# checking for feature values distribution and their value counts
cat_col = df.dtypes[df.dtypes==object].index

for col in cat_col:
    print(df[col].value_counts())
    print("==================================")
```

```python
df.head()
```

```python
CarName_price = df.groupby("CarName")["price"].mean()
CarName_price["audi"]
```

```python
df["Avg_price"] =np.where(df["CarName"], CarName_price[df['CarName']], "NA")
df['Avg_price'] = df['Avg_price'].astype('float')

conditions = [(df['Avg_price'] > 20000),(df['Avg_price'] >10000),(df['Avg_price'] > 0)]
choices = ['High', 'Medium', 'Budget']


df['Avg_price_categorical'] = np.select(conditions, choices)
```

```python
df.drop(['CarName'], axis=1,inplace=True)
df.drop(['Avg_price'], axis=1,inplace=True)
#df.drop(['price'], axis=1,inplace=True)
```

```python
df.head()
```

```python
df.doornumber.replace({"two":2,"four":4},inplace=True)
```

```python
# will removes rows which contains either hardtop or convertible or both as a carbody.
df = df[df["carbody"].str.contains("hardtop|convertible")  == False]

# will removes rows which contains 4wd as a drivewheel value
df = df[df["drivewheel"].str.contains("4wd")  == False]


# will removes rows which contains ohcf|rotor|dohcv as a enginetype value
df = df[df["enginetype"].str.contains("ohcf|rotor|dohcv")  == False]

# will removes rows which contains mfi|spfias a fuelsystem value
df = df[df["fuelsystem"].str.contains("mfi|spfi")  == False]
```

#### one hot encoding 

on features aspiration, doornumber, carbody,cylindernumber and drivewheel

```python
dummies = pd.get_dummies(df[['fueltype', 'aspiration','carbody','cylindernumber','drivewheel','enginetype','fuelsystem','Avg_price_categorical']], drop_first=True)
dummies.head()
```

```python
df= pd.concat([df.drop(['fueltype', 'aspiration','carbody','cylindernumber','drivewheel','enginetype','fuelsystem','Avg_price_categorical'],axis=1), dummies],axis=1)
df.head()
```

```python
df.drop(['enginelocation'], axis=1,inplace=True)
```

```python
#### Create A New Column combining highwaympg and citympg
df["mpg"] = (df["citympg"]+df["highwaympg"])/2
```

```python
df.drop(["citympg"],axis=1,inplace=True)
df.drop(["highwaympg"],axis=1,inplace=True)
```
